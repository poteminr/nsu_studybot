```python
import easyocr
from glob import glob
from cv2 import cv2
import matplotlib.pyplot as plt

reader = easyocr.Reader(['ru','en']);
```

```python
list_of_images = glob('assignments_images/*')
list_of_images
```

```python
image = list_of_images[2]
```

```python
result = reader.readtext(image, decoder='wordbeamsearch')
```

```python
img = cv2.imread(image)
im_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
```

```python
fig, ax = plt.subplots()

for prediction in range(len(result)):
    try:
        coord = result[prediction][0]
        im_rgb = cv2.rectangle(im_rgb, (coord[0][0], coord[0][1]), (coord[2][0], coord[2][1]), (255,0,0), 2)
    except Exception:
        pass

ax.imshow(im_rgb)
fig.set_figwidth(12)    
fig.set_figheight(12)
plt.show()
```
